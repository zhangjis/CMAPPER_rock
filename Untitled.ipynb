```python
def f_M_dm(c_z,m_z,c_m,m_m,m_start):
    # return the mass profile of the planet
    zone = c_z+m_z
    h_c = c_m/c_z
    h_m = m_m/m_z
    mass=np.zeros(zone)
    h=np.zeros(zone)
    #mass[0]=h_c
    #h[0]=h_c
    len_top=int(m_z*0.5)
    len_bot=m_z-len_top
    top_array=np.linspace(4.0,m_start,len_top)
    bot_array=np.linspace(m_start,4.0,len_bot)
    mantle_tanh=np.zeros(m_z)
    mantle_norm=np.zeros(m_z)
    mantle_tanh_sum=0.0
    for i in range(m_z):
        if i<len_bot:
            mantle_tanh[i]=0.5*(1.0+math.tanh(bot_array[i]))
        else:
            mantle_tanh[i]=0.5*(1.0+math.tanh(top_array[i-len_bot]))
        mantle_tanh_sum=mantle_tanh_sum+mantle_tanh[i]
    for i in range(m_z):
        mantle_norm[i]=mantle_tanh[i]/mantle_tanh_sum*m_m
        
    c_array=np.linspace(2.0,4.0,c_z)
    c_tanh=np.zeros(c_z)
    c_norm=np.zeros(c_z)
    c_tanh_sum=0.0
    for i in range(c_z):
        c_tanh[i]=0.5*(1.0+math.tanh(c_array[i]))
        c_tanh_sum=c_tanh_sum+c_tanh[i]
    for i in range(c_z):
        c_norm[i]=c_tanh[i]/c_tanh_sum*c_m
    mass[0]=c_norm[0]
    h[0]=c_norm[0]
    for i in range(1,zone):
        if i<c_z:
            mass[i]=mass[i-1]+c_norm[i]
            h[i]=c_norm[i]
        elif i>=c_z and i<c_z+m_z:
            #mass[i]=mass[i-1]+h_m
            #h[i]=h_m
            mass[i]=mass[i-1]+mantle_norm[i-c_z]
            h[i]=mantle_norm[i-c_z]
    return mass, h,c_norm
```

```python
# Constants
earth_mass_kg = 5.972e24  # Mass of Earth in kg
GPa_conversion = 1e-9  # Convert pressure from Pa to GPa

# File paths
path_to_folder="results_Mpl1.0_CMF0.326_time14.0_Qrad1.0_1.0_1.0_1.0_Teq255.0_Qradc0.0_eta1.0_mzmulti1.0/profile/"
file_paths = [
    path_to_folder+"StructureProfile_100.txt",
    path_to_folder+"StructureProfile_1000.txt",
    path_to_folder+"StructureProfile_2511.txt",
    path_to_folder+"StructureProfile_3981.txt",
    path_to_folder+"StructureProfile_100000.txt",
    path_to_folder+"StructureProfile_3981071.txt",
    path_to_folder+"StructureProfile_50118723.txt",
    path_to_folder+"StructureProfile_1000000000.txt",
    path_to_folder+"StructureProfile_5000000000.txt",
    path_to_folder+"StructureProfile_10000000000.txt",
    path_to_folder+"StructureProfile_14000000000.txt",]
# Functions for solidus and liquidus
def Sti_liq(P): return 5400.0 * (P / 140.0) ** 0.48
def Sti_sol(P): return 5400.0 * (P / 140.0) ** 0.48 / (1.0 - np.log(0.79))

def Fiq_liq(P): return 2022.0 + 54.21 * P - 0.34 * P**2 + 9.0747e-4 * P**3
def Fiq_sol(P): return 1621.0 + 38.415 * P - 0.1958 * P**2 + 3.8369e-4 * P**3

def T_sol_fiq(P):
    value1 = Fiq_sol(P)
    value2 = Sti_sol(P)
    return max(value1, value2) if P < 150 else value2

def T_liq_fiq(P):
    value1 = Fiq_liq(P)
    value2 = Sti_liq(P)
    return max(value1, value2) if P < 150 else value2

# Load data
def load_structure_file_with_header(filepath):
    return pd.read_csv(filepath, delim_whitespace=True, comment="#", header=None, skiprows=1)
# Reloading all files with proper headers
structure_data_corrected = {
    int(fp.split("_")[-1].split(".")[0]): load_structure_file_with_header(fp) for fp in file_paths}
# Assign correct column names
columns = [
    "radius", "pressure", "density", "gravitational acceleration", "temperature",
    "thermal expansion coefficient", "specific heat", "convective heat flux",
    "conductive heat flux", "total heat flux", "convective velocity",
    "mantle magnetic Reynolds number", "mantle viscosity", "mass", "mantle melt fraction"]
for data in structure_data_corrected.values():
    data.columns = columns
# Convert radius to cumulative mass fraction (in Earth masses)
for time, data in structure_data_corrected.items():
    data["mass_in_earth_mass"] = data["mass"] / earth_mass_kg  
# Convert pressure to GPa
for data in structure_data_corrected.values():
    data["pressure_GPa"] = data["pressure"] * GPa_conversion
for data in structure_data_corrected.values():
    data["solidus"] = data["pressure_GPa"].apply(T_sol_fiq)
    data["liquidus"] = data["pressure_GPa"].apply(T_liq_fiq)
# Determine appropriate time units for legend
def format_time(time):
    if time < 1e3:
        return f"{round(time)} years"
    elif time < 1e6:
        return f"{round(time / 1e3)} kyr"
    elif time < 1e9:
        return f"{round(time / 1e6)} Myr"
    else:
        return f"{round(time / 1e9)} Gyr"
# Use a colormap (viridis) for consistent coloring
time_keys = sorted(structure_data_corrected.keys())
cmap = cm.get_cmap("viridis", len(time_keys))
colors = cmap(np.linspace(0, 1, len(time_keys)))
```
