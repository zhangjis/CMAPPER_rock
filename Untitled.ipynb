```python
def f_M_dm(c_z,m_z,c_m,m_m,m_start):
    # return the mass profile of the planet
    zone = c_z+m_z
    h_c = c_m/c_z
    h_m = m_m/m_z
    mass=np.zeros(zone)
    h=np.zeros(zone)
    #mass[0]=h_c
    #h[0]=h_c
    len_top=int(m_z*0.5)
    len_bot=m_z-len_top
    top_array=np.linspace(4.0,m_start,len_top)
    bot_array=np.linspace(m_start,4.0,len_bot)
    mantle_tanh=np.zeros(m_z)
    mantle_norm=np.zeros(m_z)
    mantle_tanh_sum=0.0
    for i in range(m_z):
        if i<len_bot:
            mantle_tanh[i]=0.5*(1.0+math.tanh(bot_array[i]))
        else:
            mantle_tanh[i]=0.5*(1.0+math.tanh(top_array[i-len_bot]))
        mantle_tanh_sum=mantle_tanh_sum+mantle_tanh[i]
    for i in range(m_z):
        mantle_norm[i]=mantle_tanh[i]/mantle_tanh_sum*m_m
        
    c_array=np.linspace(2.0,4.0,c_z)
    c_tanh=np.zeros(c_z)
    c_norm=np.zeros(c_z)
    c_tanh_sum=0.0
    for i in range(c_z):
        c_tanh[i]=0.5*(1.0+math.tanh(c_array[i]))
        c_tanh_sum=c_tanh_sum+c_tanh[i]
    for i in range(c_z):
        c_norm[i]=c_tanh[i]/c_tanh_sum*c_m
    mass[0]=c_norm[0]
    h[0]=c_norm[0]
    for i in range(1,zone):
        if i<c_z:
            mass[i]=mass[i-1]+c_norm[i]
            h[i]=c_norm[i]
        elif i>=c_z and i<c_z+m_z:
            #mass[i]=mass[i-1]+h_m
            #h[i]=h_m
            mass[i]=mass[i-1]+mantle_norm[i-c_z]
            h[i]=mantle_norm[i-c_z]
    return mass, h,c_norm
```
